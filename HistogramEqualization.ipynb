Histogram Equalization redistributes the pixel intensities of an image to make its histogram more uniform, thereby improving its contrast.


Histogram Equalization using pre-defined function

```python
import cv2
import matplotlib.pyplot as plt

# Read the input grayscale image
image = cv2.imread("/content/car1.png", cv2.IMREAD_GRAYSCALE)

# Apply Histogram Equalization
equalized_image = cv2.equalizeHist(image)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.show()
```

Histogram Equilization using user defined function

Histogram equalization is implemented in the function histogram_equalization, which first computes the histogram of the image, then calculates the cumulative distribution function (CDF) to normalize it, and finally maps the pixel values of the input image using the normalized CDF to achieve enhanced contrast.

```python
import numpy as np

def histogram_equalization(image):
    # Compute histogram
    histogram = np.zeros(256, dtype=int)
    for pixel_value in image.flatten():
        histogram[pixel_value] += 1

    # Compute cumulative distribution function (CDF)
    cdf = np.cumsum(histogram)

    # Normalize CDF to range [0, 255]
    cdf_normalized = ((cdf - cdf.min()) * 255) / (cdf.max() - cdf.min())

    # Map pixel values using the normalized CDF
    equalized_image = cdf_normalized[image]

    return equalized_image.astype(np.uint8)


equalized_image_self = histogram_equalization(image)
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image_self, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.show()

```
