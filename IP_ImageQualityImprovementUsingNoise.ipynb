```python
import cv2
import matplotlib.pyplot as plot
import random
import numpy as np
```

1. Read the input image.


```python
img = cv2.imread('Tvisha.jpg')
image1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plot.imshow(image1)
plot.show()
```

```python
gray_image = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
plot.imshow(gray_image, cmap='gray')
plot.show()
```

2. Add noise to the input image. (Make use of Use imnoise( ) ).

```python
noise_image = []
random_value= random.randint(0,100)
temp = []
for i in gray_image:
  for pixel in i:
    if pixel < random_value:
      temp.append(0)
    elif pixel >= random_value:
      temp.append(255)
    else:
      temp.append(pixel)

  noise_image.append(temp)
  temp = []
```

```python
plot.imshow(noise_image, cmap='gray')
plot.show()
```

```python
lpf = np.ones((3, 3), np.float32) / 9
hpf = np.array([[-2, -2, -2],[-2, 16, -2],[-2, -2, -2]])
median_size = 3


```

```python
lpf_output = cv2.filter2D(gray_image, -1, lpf)
hpf_output = cv2.filter2D(gray_image, -1, hpf)
median_res = cv2.medianBlur(gray_image, median_size)
```

```python
plot.imshow(lpf_output, cmap='gray', label='LPF Img')
plot.show()
plot.imshow(hpf_output, cmap='gray', label='HPF Img')
plot.show()
plot.imshow(median_res, cmap='gray', label='Median Img')
plot.show()
```

```python
noise_image_final= np.array(noise_image, dtype=np.uint8)
lpf_noise_output = cv2.filter2D(noise_image_final, -1, lpf)
hpf_noise_output = cv2.filter2D(noise_image_final, -1, hpf)
median_noise_res = cv2.medianBlur(noise_image_final, median_size)
```

```python
plot.imshow(lpf_noise_output, cmap='gray')
plot.show()
plot.imshow(hpf_noise_output, cmap='gray')
plot.show()
plot.imshow(median_noise_res, cmap='gray')
plot.show()
```
